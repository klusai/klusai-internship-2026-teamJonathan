```python
#Create an API client
from anthropic import Anthropic

client = Anthropic()

model = "claude-haiku-4-5-20251001"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    asssistant_message = {"role": "assistant", "content": text}
    messages.append(asssistant_message)

def chat(messages, system=None, termperature=0.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": termperature
    }
    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text

def stream_chat(messages, system=None, termperature=0.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": termperature,
    }
    if system:
        params["system"] = system

    with client.messages.stream(**params) as stream:
        for text in stream.text_stream:
            print(text, end="")
    return stream.get_final_text()
```

```python
#empty message list
messages = []

while True:
    user_input = input("You: ")
    print("You: ", user_input)
    add_user_message(messages, user_input)
    response = stream_chat(messages)
    print('\n')
    add_assistant_message(messages, response)
```
