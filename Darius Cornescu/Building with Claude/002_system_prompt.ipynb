```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content":text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content":text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )

    return message.content[0].text
```

```python
messages = [] 
add_user_message(messages, "How can I solve x^2 - 4x + 4 = 0 for x?")
answer = chat(messages)
print(answer)
```

### Added -- *system prompt* --

```python
def chat(messages, system = None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    if system: params["system"] = system
    message = client.messages.create(**params)

    return message.content[0].text
```

```python
messages = [] 
system_prompt = """ 
    You're a patient math tutor. Do NOT give the direct solving solution, instead guide them to a solution step by step.
    If they dont get something try explain in simpler terms.
"""
add_user_message(messages, "How can I solve x^2 - 4x + 4 = 0 for x?")
answer = chat(messages)
print(answer)
```

```python
answer = chat(messages, system=system_prompt)
print(answer)
```
