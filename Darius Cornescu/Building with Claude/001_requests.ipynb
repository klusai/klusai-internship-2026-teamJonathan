```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What's biomechanics? And how can it influence performace in track and field? Answer in 2 sentences"
        }
    ]
)
```

## Helper functinons for creating a conversation

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content":text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content":text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )

    return message.content[0].text
```

```python
messages = [] # storing conversation list

# Added in the initial user question
add_user_message(messages, "Define biomechanics in one sentence")

# Passed the list of messages in chat
answer = chat(messages)

# Now we need to add the answer to the list
add_assistant_message(messages, answer)

# Follow-up question
add_user_message(messages, "Write another sentence")

# Get the follow-up response
final_answer = chat(messages)
```
